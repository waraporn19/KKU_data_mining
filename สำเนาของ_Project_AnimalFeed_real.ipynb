1. ถ้ามีค่าตั้งแต่ 10 ตันขึ้นไปจะให้เป็น ' Yes_ชื่อคอลั่มใหม่ที่จัดกลุ่ม ' คือมีการนำเข้าและส่งออก อาหารสัตว์

2. ถ้ามีค่าน้อยกว่า 10 ตันจะให้เป็น ' No_ชื่อคอลั่มใหม่ที่จัดกลุ่ม ' คือไม่มีการนำเข้าและส่งออกอาหารสัตว์

###เพิ่มคอลั่มใหม่ในตาราง Import_df

```
Import_df.columns
```

```
Im_mains_bucket = ['No_Im_main','Yes_Im_main'] #กำหนดข้อมูลที่นำไปใส่ในคอลั่มใหม่
Import_df['Im_main_group'] = pd.cut(Import_df.Import_of_main_raw_materials,#เพิ่มคอลั่มใหม่ Im_main_groupในตาราง Import_df 
                            bins=[0,10,10000000], #กำหนดช่วงของข้อมูล เช่น 0,10 คือเริ่มจาก 0 ไปถึง 9.99  โดยจะให้ค่าที่อยู่ในช่วงนี้แทนด้วย No_Im_main
                            labels=Im_mains_bucket,#ข้อมูลที่จะเติมใน Im_mains_bucket มาจาก No_Im_main และ Yes_Im_main ที่กำหนดไว้
                            right=False).astype(str).astype(CategoricalDtype(Im_mains_bucket,ordered=True))
```

```
Im_mixed_bucket = ['No_Im_mixed','Yes_Im_mixed']
Import_df['Im_mixed_group'] = pd.cut(Import_df.Impor_the_mixed,
                            bins=[0,10,10000000],
                            labels=Im_mixed_bucket,
                            right=False).astype(str).astype(CategoricalDtype(Im_mixed_bucket,ordered=True))
```

```
Im_petsup_bucket = ['No_Im_petsup','Yes_Im_petsup']
Import_df['Im_petsup_group'] = pd.cut(Import_df.Import_of_pet_supplement,
                            bins=[0,10,10000000],
                            labels=Im_petsup_bucket,
                            right=False).astype(str).astype(CategoricalDtype(Im_petsup_bucket,ordered=True))
```

```
Im_other_bucket = ['No_Im_other','Yes_Im_other']
Import_df['Im_other_group'] = pd.cut(Import_df.Import_other_animal_feed,
                            bins=[0,10,10000000],
                            labels=Im_other_bucket,
                            right=False).astype(str).astype(CategoricalDtype(Im_other_bucket,ordered=True))
```

##Assosiation ตาราง Animalfeed_mg

**Assosiation** เป็นเทคนิคหนึ่งของ Data Mining คือการค้นหาความสัมพันธ์ของข้อมูล
จากข้อมูลขนาดใหญ่ที่มีอยู่เพื่อจะนำไปหารูปแบบที่เกิดขึ้นบ่อยๆ
(frequent pattern) และใช้ในการวิเคราะห์ความสัมพันธ์หรือทำนาย
ปรากฏการณ์ต่างๆ 

**Apyori**    เป็นการหาความสัมพันธ์ของข้อมูลจะทำการไล่ transaction ครั้งเดียวในแต่ละระดับชั้น ในการตรวจดูว่า transaction นั้นบรรจุเซตไอเท็มใดบ้าง เพื่อความรวดเร็วจะเก็บเซตไอเท็มในแต่ละระดับชั้นทั้งหมดไว้ในโครงสร้าง Hash Tree (ต้นไม้ของข้อมูลที่ถูกสรุปแล้ว) จุดเด่นความเร็วของการค้นหาไอเท็มเซตที่ปรากฏบ่อย ด้วยการละเว้นการพิจารณาไอเท็มเซตที่ปรากฏซ้าด้วยที่ Confidence ต่ำกว่าเกณฑ์



**โดยกำหนดให้**
**Transection** คือ ประเทศนำเข้าและส่งออกอาหารสัตว์มีทั้งหมด 36 ประเทศ

**item** คือ ปริมาณนำเข้าและส่งออกอาหารสัตว์

**ค่าสนับสนุน (Support)** คือ เปอร์เซ็นต์ของจำนวน Item sets ทั้งหมดที่เกิดขึ้นในฐานข้อมูล

**ค่าความเชื่อมั่น (Confidence)** คือ เปอร์เซ็นต์ของจำนวน Item sets ทั้งหมดที่เกิดขึ้นในฐานข้อมูลต่อจำนวน Item sets ที่เกิดขึ้น

**ค่าสหสัมพันธ์ (Lift)** คือค่าที่บ่งบอกว่ากฎมีความสัมพันธ์กัน หรือไม่ โดยถ้าค่าสหสัมพันธ์(Lift) มีค่ามากกว่า 1 แสดงว่ากฎทั้งสองมีความสัมพันธ์กัน แต่ถ้ามีค่าน้อยกว่า 1 แสดงว่า กฎไม่ความสัมพันธ์กันหรือขึ้นต่อกัน (Independent)
